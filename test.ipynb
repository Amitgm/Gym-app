```python
from langchain.docstore.document import Document

# List to hold the Document objects
langchain_documents = []

# Iterate through the sample data and create Document objects
for item in document_data:
    # Formulate the page_content string
    page_content = (
        f"Title: {item['Title']}\n"
        f"Type:{item['Type']}\n"
        f"BodyPart: {item['BodyPart']}\n"
        f"Desc: {item['Desc']}\n"
        f"Equipment: {item['Equipment']}\n"
    )
    
    # Create the metadata dictionary
    metadata = {"Level": item['Level']}
    
    # Create the Document object
    doc = Document(page_content=page_content, metadata=metadata)
    
    # Add the Document to our list
    langchain_documents.append(doc)

# Now, langchain_documents contains your list of Document objects
# You can print them to verify (optional)
for i, doc in enumerate(langchain_documents):
    print(f"--- Document {i+1} ---")
    print(f"Page Content:\n{doc.page_content}")
    print(f"Metadata: {doc.metadata}\n")
```

```python
# Build prompt
from langchain_core.prompts import PromptTemplate     
from langchain_core.output_parsers import StrOutputParser                            # to transform the output of an LLM into a more usable format
from langchain.schema.runnable import RunnableParallel, RunnablePassthrough        
```

```python
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory
from langchain_community.chat_models import ChatOpenAI

prompt = PromptTemplate.from_template("""
                                          
You are a helpful fitness assistant. Use the following context to answer the question The {level} level is provided for you to get a better idea on how to answer the question
                                .
If you don't know the answer, just say that you don't know, don't try to make up an answer.
Context:
{context}

Chat History:
{history}

Question:
{question}

Answer:
""")

memory = ConversationBufferMemory(memory_key="history", input_key="question", return_messages=True)

chain = LLMChain(
    llm=ChatOpenAI(temperature=0.5,model_name="gpt-4o"),
    prompt=prompt,
    memory=memory
)

# Pass your context manually from retrieved documents
context = "\n\n".join([doc.page_content for doc in results])

response = chain.run({"context": context, "question": "Gym Dumbells for legs?","level":"Beginner"})
print(response)
```
